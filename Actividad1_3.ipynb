##a) Eliminar la fila “0” de “Unnamed” y Eliminar la columna “0” y la columna “Referencia”

```python
#Eliminar primera fila o filas innecesarias
filtroa=pd.read_excel('Inmuebles.xlsx', skiprows=1)
filtroa
```

```python
#Eliminar columnas innecesarias
#axis= 1 para columnas
#axis= 0 para filas
filtroa1= filtroa.drop(['Referencia'], axis=1)
filtroa1
```

```python
#Convertir archivo filtrado a csv
filtroa.to_csv("Filtro1.a.csv")
```

##b) Todos las registros de “Fecha Alta : igual a 2007-04-29 00:00:00 y a 2004-01-02 00:00:00

```python
#Filtro or
filtrob = filtroa1[(filtroa1['Fecha_Alta'] == '2007-04-29 00:00:00') | (filtroa1['Fecha_Alta'] == '2004-01-02 00:00:00')]
filtrob
```

```python
#Convertir archivo filtrado a csv
filtrob.to_csv("Filtro1.b.csv")
```

```python
#Convertir archivo filtrado a CSV
from google.colab import files

files.download("Filtro1.b.csv")
```

##c) Todos los registros de “ Tipo: parking, local, industrial y piso”

```python
#Filtro or
filtroc = filtroa1[(filtroa1['Tipo'] == 'Parking') | (filtroa1['Tipo'] == 'Local')|(filtroa1['Tipo'] == 'Industrial') | (filtroa1['Tipo'] == 'Piso')]
filtroc
```

```python
#Convertir archivo filtrado a csv
filtroc.to_csv("Filtro1.c.csv")
```

```python
#Convertir archivo filtrado a CSV
from google.colab import files

files.download("Filtro1.c.csv")
```

##d) Todos los registros con “Provincia” igual a “Girona” y con “Superficie” mayor a “100”

```python
#Filtro or
filtrod = filtroa1[(filtroa1['Provincia'] == 'Girona') | (filtroa1['Superficie'] > 100)]
filtrod
```

```python
#Convertir archivo filtrado a csv
filtrod.to_csv("Filtro1.d.csv")
```

```python
#Convertir archivo filtrado a CSV
from google.colab import files

files.download("Filtro1.d.csv")
```

##e) Todos los registros con “Precio Venta: mayor a 1000,000” o “vendedor” igual a “Carmen o Joaquin”

```python
#Filtro or
filtroe = filtroa1[(filtroa1['Precio Venta'] > 1000000) | (filtroa1['Vendedor']== 'Carmen')| (filtroa1['Vendedor']== 'Joaquin')]
filtroe
```

```python
#Convertir archivo filtrado a csv
filtroe.to_csv("Filtro1.e.csv")
```

```python
#Convertir archivo filtrado a CSV
from google.colab import files

files.download("Filtro1.e.csv")
```

##f) Todos los registros que no sean de “Provincia : Barcelona“

```python
#Convertir archivo filtrado a CSV
from google.colab import files

files.download("Filtro1.f.csv")
```
