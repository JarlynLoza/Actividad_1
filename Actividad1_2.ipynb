##1-CVE_CLPV      (valores: 1000 a 2000)

```python
filtro1=df[(df["CVE_CLPV"]>=1000)& (df["CVE_CLPV"]<=2000)]
filtro1
```

```python
#Convertir tipo de dato de columna de tipo Obj a Float
df["CVE_CLPV"]=df['CVE_CLPV'].astype(float)
```

```python
df.CVE_CLPV=df.CVE_CLPV.replace({'M':'0'}, regex=False)
```

```python
#Convertir archivo filtrado a csv
filtro1.to_csv("Filtro1.1csv")
```
